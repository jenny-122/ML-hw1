```python
#HW1 Part 1
#given data

import numpy as np
import matplotlib.pyplot as plt

#full set

#temperature = x
x = [52, 55, 50, 70, 75, 72, 80, 82, 85, 97, 95, 90, 105, 108, 100]

#yield = y
y = [3.3, 2.8, 2.9, 2.3, 2.6, 2.1, 2.5, 2.9, 2.4, 3.0, 3.1, 2.8, 3.3, 3.5, 3.0]

#validation pts = {1, 4, 7, 10, 13}
valid_index_pts = [1, 4, 7, 10, 13]
```

```python
m=1 # degree of polynomial regression function
arrM = np.array([1, 3, 5, 10, 15]) # degree values for polyfit, not poly1d
for m in arrM:
    model = np.polyfit(x, y, m) 
    mod = np.poly1d(model) #returns a polynomical function
    print(model)
    print('m =', m)
    print('model =', model)
    print('')
    #part 2: plot the data
    plt.scatter(x, y)
    line = np.linspace(40, 110, 100)
    plt.plot(line, mod(line))
    plt.show()
    #part 3: find the coefficients
    print('coefficents=', mod.c)
    print('')
```
