```python
#HW1 Part 1
#given data

import numpy as np
import matplotlib.pyplot as plt

#full set

#temperature = x
x = [52, 55, 50, 70, 75, 72, 80, 82, 85, 97, 95, 90, 105, 108, 100]

#yield = y
y = [3.3, 2.8, 2.9, 2.3, 2.6, 2.1, 2.5, 2.9, 2.4, 3.0, 3.1, 2.8, 3.3, 3.5, 3.0]

#validation index = {1, 4, 7, 10, 13}; starts @ 1 not 0
valid_index_pts = [1, 4, 7, 10, 13]
arr_valid_index_pts = []
for i in valid_index_pts:
    arr_valid_index_pts.append(i-1)
print(arr_valid_index_pts)

#training index = {2, 3, 5, 6, 8, 9, 11, 12, 14, 15}; starts @ 1 not 0
train_index_pts = [2, 3, 5, 6, 8, 9, 11, 12, 14, 15]
arr_train_index_pts = []
for i in train_index_pts:
    arr_train_index_pts.append(i-1)
print(arr_train_index_pts)
```

```python
#training set
x_train = []
y_train = []
for i in arr_train_index_pts:
    x_train.append(x[i])
    y_train.append(y[i])
print("x_train = ", x_train)
print("y_train = ", y_train)

#validation set
x_valid = []
y_valid = []
for i in arr_valid_index_pts:
    x_valid.append(x[i])
    y_valid.append(y[i])
print('x_valid =', x_valid)
print('y_valid =', y_valid)

print(arr_train_index_pts)
```

```python
def squared_mean_error(v, p): #validation, prediction arrays
    diffArr = []
    sqrDiff = (v - p)**2
    diffArr.append(sqrDiff)
    return np.mean(diffArr)
```
